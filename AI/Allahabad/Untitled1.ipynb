```python
import pandas as pd
df = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
df['target'] = iris.target
```

```python
df['flower'] = df.target.apply(lambda x : iris.target_names[x])
```

```python
df
```

```python
df0 = df[:50]
df1 = df[50:100]
df2= df[100:150]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'])
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'])
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'])
```

```python
plt.scatter(df0['petal length (cm)'], df0['sepal width (cm)'])
plt.scatter(df1['petal length (cm)'], df1['sepal width (cm)'])
# plt.scatter(df2['petal length (cm)'], df2['sepal width (cm)'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['target', 'flower'], axis = 'columns')
y = df.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
```

```python
model
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.predict([[5,5,5,5]])
```

```python
y_pred = model.predict()
```

```python
y_pred
```

```python
model.score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
import seaborn as sns
plt.figure(figsize=(5,5))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Original')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
