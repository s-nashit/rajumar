```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(4,4))
```

```python
#binary activation : output layer
def Sigmoid_b(x):
    return(1/(1+np.exp(-x)))

x = np.linspace(-5,5)
plt.plot(x, Sigmoid_b(x))
plt.title('Activation - binary sigmoid function')
plt.show()
```

```python
def TanH(x):
    return np.tanh(x)

plt.plot(x, TanH(x))
plt.title('Activation - tanh function')
plt.show()
```

```python
def RectifiedLinearUnit(x):
    list=[]
    for i in x:
        if i>=0:
            list.append(i)
        else:
            list.append(0)
    return list
x = np.linspace(-5,5)
plt.plot(x, RectifiedLinearUnit(x))
plt.title('Activation - reLU : prediction')
plt.show()
```

```python
def LeakyreLu(x):
    list = []
    for i in x:
        if i>=0:
            list.append(i)
        else:
            list.append(0.01*i)
    return list
x = np.linspace(-5,5)
plt.plot(x, LeakyreLu(x))
plt.title('Activation - leakyRELU')
plt.show()
```

```python
def softmax(x):
    return (np.exp(x)/np.sum(np.exp(x)))

plt.plot(x, softmax(x))
plt.title('Activation - Softmax')
plt.show()
```
