```python
a = pd.read_csv('p.csv')
```

```python
a.head()
```

```python
x = a['Name']
y = a['Marks']
```

```python
plt.plot(x,y,'#05A4B3',lw=3)
plt.ylabel('Marks')
plt.xlabel('Name')
plt.title('ADIT Class', {'color':'red', 'size':15,'family':'serif'})
plt.grid(axis='both', mouseover=True)
```

```python
plt.figure(figsize=(12,3))
plt.barh(x,y)
plt.ylabel('Marks')
plt.xlabel('Name')
plt.title('ADIT Class', {'color':'red', 'size':15,'family':'serif'})
plt.grid(axis='both')
```
