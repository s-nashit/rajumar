```python
#importing libraries
import pandas as pd
#read dataset
df =pd.read_csv('abc.csv')
```

```python
#check null values
df.isnull()
```

```python
#check for the info
df.info()
```

```python
#null values dropped
df.dropna(inplace=True)
```

```python
# cleaned data
df
```

```python
# cleaned data info
df.info()
```

```python
#preparing data
x = df.drop('TenYearCHD', axis='columns')
```

```python
y = df['TenYearCHD']
```

```python
#importing libraries
from  sklearn.model_selection import train_test_split
```

```python
#training & testing data preparation
x_train, x_test, y_train, y_test= train_test_split(x, y, train_size=0.65)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# model building
model=LogisticRegression()
```

```python
#model training
model.fit(x_train, y_train)
```

```python
# predicting data for test data
y_pred = model.predict(x_test)
```

```python
#predicted data
y_pred
```

```python
#probablity
df2 = model.predict_proba(x_test)
df3 = pd.DataFrame(df2)
df3.to_csv('diabe.csv')
```

```python
#model score
model.score(x_test,y_test)
```

```python
#predicted & original values preparation for human 
df1 = pd.DataFrame({'original': y_test, 'predicted' : y_pred})
```
